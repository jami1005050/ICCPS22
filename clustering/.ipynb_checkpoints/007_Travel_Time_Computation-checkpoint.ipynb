# Load all speed data (DF_All)

```python
from copy import deepcopy
df_overall = []
to_months = 2
for month in tqdm(range(1, to_months)):
    fp = os.path.join(speed_dir, f'ALL_5m_DF_2019_{month}_1.gzip')
    DF_1 = Read_DF(DF_All = fp, Reading_Tag = 'DF_All', MetaData = {})

    fp = os.path.join(speed_dir, f'ALL_5m_DF_2019_{month}_15.gzip')
    DF_15 = Read_DF(DF_All = fp, Reading_Tag = 'DF_All', MetaData = {})

    df_month_all = pd.concat([DF_1.reset_index(), DF_15.reset_index()])
#     df_month_all = df_month_all[['time_local', 'XDSegID', 'speed_mean', 'average_speed', 'congestion_mean', 'Total_Number_Incidents']]
    df_month_all = df_month_all.set_index('time_local')
    df_overall.append(df_month_all)
```

# Generating routes that pass through incidents
* These are generated randomly and i did not use any OD probabilities to compute the locations, can be improved like with the osaka data
* type: `list of dicts`
* format: `[{'route_index': int, 'route': list, 'incident_dt': pd.DateTime, 'segment_with_incident': int}]`

```python
time_start = time.time()

number_of_routes = 1
ii = 0
routes_list = []

pbar = tqdm(total=number_of_routes)

while len(routes_list) < number_of_routes:
    sample_incident = df_incidents_gt.sample(1)
    sample_incident_ts = sample_incident.index[0]
    if sample_incident_ts.hour < 6:
        continue
    sample_incident_segment = sample_incident.XDSegID.values[0]
    
    # Pick a random segment within the area
    r = None
    while not r:
        origin = random.choice(list(L_XDSegID.nodes))
        destin = random.choice(list(L_XDSegID.nodes))
        try:
            r = nx.shortest_path(L_XDSegID, origin, destin)
            if sample_incident_segment not in r:
                r = None
            else:
                route_dict = {'route_index': ii,
                              'route': r,
                              'incident_dt': sample_incident_ts,
                              'segment_with_incident': sample_incident_segment}
                routes_list.append(route_dict)
                ii += 1
                pbar.update(1)
        except:
            pass

pbar.close()

time_elapsed = time.time() - time_start
print(f'Total time to generate routes: {time_elapsed} s')
```

```python
all_routes_df_list = []
for r_dict in tqdm(routes_list):
    sample_incident_ts = r_dict['incident_dt']
    route = r_dict['route']
    route_index = r_dict['route_index']
    
    segment_incident = r_dict['segment_with_incident']

    incident_month = sample_incident_ts.month
    df_month_all = df_overall[incident_month - 1]
    
    route_dict = {'length_miles':[], 'segments': [], 'dirty_speed_mph': [], 'incident': [], 'clean_speed_mph': []}
    failed = False
    for _r in route:
        seg_leng_miles = inrix_df[inrix_df['XDSegID'] == _r]['Miles'].values[0]
        route_dict['length_miles'].append(seg_leng_miles)
        route_dict['segments'].append(_r)

        if _r == segment_incident:
            route_dict['incident'].append(1)
        else:
            route_dict['incident'].append(0)

        before = sample_incident_ts - pd.Timedelta('1h')
        after  = sample_incident_ts + pd.Timedelta('1h')
        for_clean_df = df_month_all[(df_month_all.index.day == sample_incident_ts.day) &
                                    (df_month_all['XDSegID'] == _r)]
        
        dirty_speed = for_clean_df[before:after].speed_mean.mean()
        route_dict['dirty_speed_mph'].append(dirty_speed)

        try:
            a = cleaning_data.clean_speed_data_basic(for_clean_df, sample_incident_ts, window=1)
            clean_speed = a[before:after].speed_mean.mean()
            route_dict['clean_speed_mph'].append(clean_speed)
        except:
            failed = True
            break

    #     print(seg_leng_miles)
    if not failed:
        print("Success")
        route_df = pd.DataFrame(route_dict)
        route_df = route_df.dropna()
        route_df['r_index'] = route_index
        all_routes_df_list.append(route_df)
```

## Computation proper

```python
segment_windows = 2

route_results_list = []
pbar = tqdm(all_routes_df.r_index.unique())
for route in pbar:
    pbar.set_description(f'Route {route}')
    route_df = all_routes_df[all_routes_df['r_index'] == route]
    if route_df.index[route_df['incident'] == 1].empty:
        continue
    incident_segment = route_df.index[route_df['incident'] == 1].tolist()[0]

    # during and N segments after
    select = route_df.index.isin(range(incident_segment, incident_segment + segment_windows))
    
    route_df.loc[~select, 'used_speed_mph'] = route_df['dirty_speed_mph']
    route_df.loc[select, 'used_speed_mph'] = route_df['clean_speed_mph']
    
    route_df['used_travel_time'] = route_df['length_miles'] / route_df['used_speed_mph']
    route_df['dirty_travel_time'] = route_df['length_miles'] / route_df['dirty_speed_mph']
    used_tt  = route_df['used_travel_time'].sum() * 60
    dirty_tt = route_df['dirty_travel_time'].sum() * 60
    
#     # We assume that dirty_tt takes longer
    saved_tt = dirty_tt - used_tt

    route_results_dict = {}
    route_results_dict['r_index'] = route
    route_results_dict['speed_diff_mph'] = route_df['used_speed_mph'].mean() - route_df['dirty_speed_mph'].mean()
    route_results_dict['tt_diff_minutes'] = saved_tt
    route_results_list.append(route_results_dict)
#     break
    
route_results_df = pd.DataFrame(route_results_list)
```

```python
# Duplicating the results N times for the random chances
RUNS = 5
route_results_df = pd.concat([route_results_df] * RUNS, ignore_index=True)
```

```python
print(route_results_df.shape)
route_results_df.tt_diff_minutes.sum()
```

```python
# Using the different probabilities
for i, chance in enumerate(chance_to_detect):
    route_results_df[i] = [random.uniform(0, 1) <= chance for _ in range(route_results_df.shape[0])]
```
